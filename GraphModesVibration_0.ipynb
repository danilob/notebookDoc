### Gráficos dos Modos de Vibração : Modo 0
#### Comparação ângulos nos eixos X,Y e Z e Posição relativa ao Torso nos eixos X,Y e Z das juntas
    

- Ângulo no eixo X

```python
import pandas as pd
from datetime import datetime
import csv
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
headers = ['Time','Pos_x','Pos_y','Pos_z','Ang_x','Ang_y','Ang_z','None']
legends = ['ThighL','ThighR','ShankL','ShankR','FootL','FootR','UpperArmL','UpperArmR','ForearmL','ForearmR']
bodies = ['thigh_leftw_','thigh_rightw_','shank_leftw_','shank_rightw_','foot_leftw_','foot_rightw_','upper_arm_leftw_','upper_arm_rightw_','forearm_leftw_','forearm_rightw_']

mode = 0 #atibua modo de vibração no intervalo de 0 a 29

j = 0
for i in range(5):   
    df_l = pd.read_csv('data/'+bodies[j]+str(mode)+'.csv',delimiter=";",names=headers)
    j = j+1
    df_r = pd.read_csv('data/'+bodies[j]+str(mode)+'.csv',delimiter=";",names=headers)
    j = j+1
    x = df_l['Time']
    yl = df_l['Ang_x']
    yr = df_r['Ang_x']
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 7),sharex=True,sharey=True)
    axes[0].plot(x,yl,label=legends[j-2]+'Angle X')
    axes[0].legend(frameon=False, loc='upper center', ncol=2)
    axes[1].plot(x,yr,color="#ff5733",label=legends[j-1]+'Angle X')
    axes[1].legend(frameon=False, loc='upper center', ncol=2)
    fig.tight_layout()
```
